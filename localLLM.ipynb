```python
import asyncio
import ollama
```

```python
async def call_ollama():
    # One‑shot chat with the model you pulled
    resp = await ollama.chat(
        model="gemma3:4b",
        messages=[
            {"role": "system", "content": "You are a helpful assistant."},
            {"role": "user",   "content": "Explain quantum entanglement in a sentence."}
        ]
    )
    # resp is a dict that mirrors the raw Ollama JSON response
    print("\n=== Answer ===")
    print(resp["message"]["content"])
```

```python
call_ollama()
```
